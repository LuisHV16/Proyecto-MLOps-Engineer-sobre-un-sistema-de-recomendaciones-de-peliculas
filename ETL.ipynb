7.- Algunos campos, como belongs_to_collection, production_companies y otros (ver diccionario de datos) están anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila, ¡deberán desanidarlos para poder y unirlos al dataset de nuevo hacer alguna de las consultas de la API! O bien buscar la manera de acceder a esos datos sin desanidarlos.

```python
df_movies.head()
```

Como podemos observar tenemos columnas anidadas por lo que debemos definir algunas funciones para desanidar los datos contenidos en listas y diccionarios.

```python
#DEFINIMOS FUNCIONES PARA DESANIDAR LAS COLUMNAS ANIDADAS

def fetch_name(obj): 
    if isinstance(obj, str) and '{' in obj:
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name']);
        return L
```

```python
def fetch_name2(obj): 
    if isinstance(obj, str) and '{' in obj:
        # print(obj)
        dic = ast.literal_eval(obj)
        return dic['name']
```

```python
#APLICAMOS LAS FUNCIONES CREADAS A LAS COLUMNAS, PARA DESANIDARLAS
df_movies.loc[:, 'genres']                = df_movies['genres'].apply(fetch_name)
df_movies.loc[:, 'belongs_to_collection'] = df_movies['belongs_to_collection'].apply(fetch_name2)
df_movies.loc[:, 'production_companies']  = df_movies['production_companies'].apply(fetch_name)
df_movies.loc[:, 'production_countries']  = df_movies['production_countries'].apply(fetch_name)
df_movies.loc[:, 'spoken_languages']      = df_movies['spoken_languages'].apply(fetch_name)
```

```python
df_movies.head(10)
```

```python
#EXPRESAMOS LOS DATOS DE LAS COLUMNAS QUE TIENEN DATOS EN FORMA DE LISTAS A FORMA DE NO LISTAS

df_movies.loc[:, 'genres'] = df_movies['genres'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
df_movies.loc[:, 'production_companies'] = df_movies['production_companies'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
df_movies.loc[:, 'production_countries'] = df_movies['production_countries'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
df_movies.loc[:, 'spoken_languages'] = df_movies['spoken_languages'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
```

Para extraer el nombre del director de la pelicula de nuestra columna crew vamos a crear una nueva funcion llamada fetch_director

```python
#Creamos la función que nos ayudara a extraer el nombre del director verificando el empleo job igual a director
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
#Aplicamos nuestra funcion a la columna crew
df_unido['crew']=df_unido['crew'].apply(fetch_director)
```

```python
#Cambiaremos el nombre de nuestra columna crew a director
df_unido = df_unido.rename(columns={'crew': 'director'})
```
